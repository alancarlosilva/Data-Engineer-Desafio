# Querying JSON & Hierarchical Data com SQL

Apache Spark&trade; Databricks&reg; torna mais fácil trabalhar com dados hierárquicos, como registros JSON aninhados.

```
%fs head /FileStore/tables/data.json
```

Para expor o arquivo JSON como uma tabela, use a tabela de criação de SQL padrão usando a sintaxe apresentada abaixo:

```
%sql
CREATE TABLE IF NOT EXISTS DatabricksAnalytics
  USING json
  OPTIONS (
    path "dbfs:/FileStore/tables/data.json",
    inferSchema "true"
  )
```

Verificando a tabela criada com a função `DESCRIBE`

```
%sql
DESCRIBE DatabricksAnalytics
```

Rodando uma query para verificar o conteúdo da tabela.

Aviso:
* A coluna `totals` é um array contendo multiplos valores aninhados

```
%sql

SELECT
channelGrouping,
customDimensions,
date,
device,
fullVisitorId,
geoNetwork,
hits,
socialEngagementType,
totals,
trafficSource,
visitId,
visitNumber,
visitStartTime
FROM DatabricksAnalytics
```

Pense em dados aninhados como colunas dentro de colunas.

Por exemplo, observe a coluna `totals`.

```
%sql
SELECT totals FROM DatabricksAnalytics
```

Para retirar um subcampo específico vamos usar notação de "ponto"

```
%sql

SELECT
totals.bounces,
totals.hits,
totals.newVisits,
totals.pageviews,
--totals.screenviews, (NÃO EXISTE)
totals.sessionQualityDim,
--totals.timeOnScreen, (NÃO EXISTE)
totals.timeOnSite,
totals.totalTransactionRevenue,
totals.transactionRevenue,
totals.transactions,
--totals.UniqueScreenViews, (NÃO EXISTE)
totals.visits
FROM DatabricksAnalytics
```

CONTAGEM DE PAGEVIEWS

```
%sql

SELECT
sum(totals.pageviews) as CONTAGEM_PAGEVIEWS
FROM DatabricksAnalytics
```

NÚMERO DE SESSÕES POR USUÁRIO

```
%sql

SELECT
visitId,
fullVisitorId,
SUM(visitNumber) AS NUMERO_DE_SESSOES
FROM DatabricksAnalytics
GROUP BY visitId, fullVisitorId
LIMIT 5
```

SESSÕES DISTINTAS POR DATA

```
%sql

SELECT DISTINCT
date_format(to_date(date, 'yyyymmdd'), "MMM dd, yyyy") AS SESSAO_DATA,
fullVisitorId,
visitId
FROM DatabricksAnalytics
LIMIT 5
```

MÉDIA DE DURAÇÃO DA SESSÃO POR DATA

```
%sql

SELECT
date_format(to_date(date, 'yyyymmdd'), "MMM dd, yyyy") AS SESSAO_DATA,
MEAN(totals.timeOnSite) AS MEDIA_SESSAO_DATA_SEGUNDOS
FROM DatabricksAnalytics
GROUP BY date_format(to_date(date, 'yyyymmdd'), "MMM dd, yyyy")
LIMIT 5
```

SESSÕES DIÁRIAS POR TIPO DE BROWSER

```
%sql

SELECT
date_format(to_date(date, 'yyyymmdd'), "MMM dd, yyyy") AS SESSAO_DATA,
device.browser AS BROWSER,
count(device.browser) AS NUMERO_VEZES_USADO
FROM DatabricksAnalytics
GROUP BY date, device.browser
LIMIT 5
```
